```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
info_string = '2, 224, 224, 0, 191, 223, 1, 92, 223, 2, 158, 223, 0, 125, 223, 2, 59, 223, 2, 26, 223, 1, 224, 190, 1, 191, 190, 1, 59, 190, 2, 158, 190, 0, 125, 190, 0, 92, 190, 2, 26, 190, 0, 224, 158, 2, 191, 157, 2, 158, 157, 0, 125, 157, 1, 92, 157, 1, 26, 157, 2, 59, 157, 1, 158, 124, 1, 224, 125, 1, 92, 124, 0, 191, 124, 2, 125, 124, 0, 59, 124, 2, 26, 124, 0, 158, 92, 1, 125, 92, 0, 92, 92, 2, 59, 92, 0, 27, 92, 2, 225, 92, 2, 191, 92, 1, 158, 59, 1, 27, 59, 0, 93, 59, 2, 60, 59, 0, 225, 59, 2, 191, 59, 2, 125, 59, 1, 125, 26, 1, 225, 26, 2, 158, 26, 0, 92, 26, 0, 60, 26, 2, 27, 26, 2, 191, 26'
```

```python
info_list = []
info_string_split = info_string.split(',')

tmp_l = []
for i in range(len(info_string_split)):
    tmp_l.append(int(info_string_split[i]))
    if (i+1) % 3 == 0:
        info_list.append(tmp_l)
        tmp_l = []
```

```python
img_size = 1000
shape_size = 20

def draw_circle(img, pos):
    cv2.circle(img, pos, shape_size, (0,0,0), -1)
    return img

def draw_rectangle(img, pos):
    x,y = pos
    top_left = (x-shape_size, y-shape_size)
    bottom_right = (x+shape_size, y+shape_size)
    cv2.rectangle(img, top_left, bottom_right, (0,0,0), -1)
    return img

def draw_triangle(img, pos):
    x,y = pos
    triangle_cnt = np.array([(x, y-shape_size) , (x-shape_size, y+shape_size), (x+shape_size, y+shape_size)])
    cv2.drawContours(img, [triangle_cnt], 0, (0,0,0), -1)
    return img

def draw_cross(img, pos):
    x,y = pos
    cross_cnt = np.array([(int(x+shape_size/4), y-shape_size), (int(x+shape_size/4), int(y-shape_size/4)),
                 (x+shape_size, int(y-shape_size/4)), (x+shape_size, int(y+shape_size/4)),
                 (int(x+shape_size/4), int(y+shape_size/4)), (int(x+shape_size/4), y+shape_size),
                 (int(x-shape_size/4), y+shape_size), (int(x-shape_size/4), int(y+shape_size/4)),
                 (x-shape_size, int(y+shape_size/4)), (x-shape_size, int(y-shape_size/4)),
                 (int(x-shape_size/4), int(y-shape_size/4)), (int(x-shape_size/4), y-shape_size)])
    cv2.drawContours(img, [cross_cnt], 0, (0,0,0), -1)
    return img

res_image = np.ones((img_size, img_size, 3), np.uint8)*255

for info in info_list:
    x = int(info[1]/255*img_size)
    y = int(info[2]/255*img_size)

    if info[0] == 0:
        res_image = draw_circle(res_image, (x,y))
    elif info[0] == 1:
        res_image = draw_triangle(res_image, (x,y))
    elif info[0] == 2:
        res_image = draw_rectangle(res_image, (x,y))
    elif info[0] == 3:
        res_image = draw_cross(res_image, (x,y))

fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(res_image, cmap='gray')
```
