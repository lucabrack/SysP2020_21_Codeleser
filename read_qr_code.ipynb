```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

qr_detector = cv2.QRCodeDetector()
```

```python
def show_qr_detection(img, pts):
    pts = np.int32(pts).reshape(-1,2)
    
    for j in range(pts.shape[0]):
        cv2.line(img, tuple(pts[j]), tuple(pts[(j+1) % pts.shape[0]]), (0, 255, 0), 5)
    
    for j in range(pts.shape[0]):
        cv2.circle(img, tuple(pts[j]), 10, (0, 0, 255), -1)

def show_result(image):
    data, bbox, qr_code = qr_detector.detectAndDecode(image)

    if len(data) > 0:
        print('Decoded Data: {}'.format(data))
        show_qr_detection(image, bbox)

    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
```

```python
image = cv2.imread('Images/qrcodeTEST001.png')
show_result(image)
```

```python
image = cv2.imread('Images/macroframe00002.jpg')
show_result(image)
```
